# Эйлер

```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.loadtxt('X_out.csv')
x_t = data1[:, 0]
x = data1[:, 1]

plt.plot(x_t, x, label='X(t)')

data2 = np.loadtxt('V_out.csv')
v_t = data2[:, 0]
v = data2[:, 1]

plt.plot(v_t, v, label='V(t)')

plt.xlabel('t')
plt.ylabel('x,v')
plt.title('График функции X(t), V(t)')

plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

n = [10001, 1001, 101, 11]
d = [[0] * len(n)]

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs = axs.ravel()

for i in range(len(n)):
    data1 = np.loadtxt("V2_out.csv")
    vt1 = data1[:n[i], 0]
    v1 = data1[:n[i], 1]

    data2 = np.loadtxt("X2_out.csv")
    xt1 = data2[:n[i], 0]
    x1 = data2[:n[i], 1]
    
    axs[i].plot(vt1, v1)
    axs[i].plot(xt1, x1)
    axs[i].set_title("Graph ")
    
    with open("X2_out.csv","r") as f:
        lines = f.readlines()
    with open("X2_out.csv","w") as f:
        f.writelines(lines[n[i]:])
    with open("V2_out.csv","r") as f:
        lines = f.readlines()
    with open("V2_out.csv","w") as f:
        f.writelines(lines[n[i]:])    
    

# Отображение графиков
plt.show()
```
