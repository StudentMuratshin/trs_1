# Эйлер

```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.loadtxt('X_out.csv')
x_t = data1[:, 0]
x = data1[:, 1]

plt.plot(x_t, x, label='X(t)')

data2 = np.loadtxt('V_out.csv')
v_t = data2[:, 0]
v = data2[:, 1]

plt.plot(v_t, v, label='V(t)')

plt.xlabel('t')
plt.ylabel('x,v')
plt.title('График функции X(t), V(t)')

plt.legend()

plt.show()
```

# Адамс

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файлов
data1 = np.loadtxt("0V_out.csv")
vt1 = data1[:, 0]
v1 = data1[:, 1]

data2 = np.loadtxt("0X_out.csv")
xt2 = data2[:, 0]
x2 = data2[:, 1]

data3 = np.loadtxt("1V_out.csv")
vt3 = data3[:, 0]
v3 = data3[:, 1]

data4 = np.loadtxt("1X_out.csv")
xt4 = data4[:, 0]
x4 = data4[:, 1]

data5 = np.loadtxt("2V_out.csv")
vt5 = data5[:, 0]
v5 = data5[:, 1]

data6 = np.loadtxt("2X_out.csv")
xt6 = data6[:, 0]
x6 = data6[:, 1]

data7 = np.loadtxt("3V_out.csv")
vt7 = data7[:, 0]
v7 = data7[:, 1]

data8 = np.loadtxt("3X_out.csv")
xt8 = data8[:, 0]
x8 = data8[:, 1]

# Создание 4 графиков
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs = axs.ravel()
# Добавление на каждый график двух графиков с координатами из двух файлов
axs[0].plot(vt1, v1)
axs[0].plot(xt2, x2)
axs[0].set_title("Graph 1")

axs[1].plot(vt3, v3)
axs[1].plot(xt4, x4)
axs[1].set_title("Graph 2")

axs[2].plot(vt5, v5)
axs[2].plot(xt6, x6)
axs[2].set_title("Graph 3")

axs[3].plot(vt7, v7)
axs[3].plot(xt8, x8)
axs[3].set_title("Graph 4")

# Отображение графиков
plt.show()
```
