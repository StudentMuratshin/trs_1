```python
n = [11, 101, 1001, 10001]
d = [[0] * len(n)]

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs = axs.ravel()

for i in range(len(n)):
    data1 = np.loadtxt("V1_out.csv")
    vt1 = data1[:n[i], 0]
    v1 = data1[:n[i], 1]

    data2 = np.loadtxt("X1_out.csv")
    xt1 = data2[:n[i], 0]
    x1 = data2[:n[i], 1]
    
    axs[i].plot(vt1, v1, label=f'v(t), h={10/(n[i]-1)}')
    axs[i].plot(xt1, x1, label=f'x(t), h={10/(n[i]-1)}')
    axs[i].set_title(f'h={10/(n[i]-1)}')
    axs[i].legend(loc='upper right')
    
    with open("X1_out.csv","r") as f:
        lines = f.readlines()
    with open("X1_out.csv","w") as f:
        f.writelines(lines[n[i]:])
    with open("V1_out.csv","r") as f:
        lines = f.readlines()
    with open("V1_out.csv","w") as f:
        f.writelines(lines[n[i]:])    
    

# Отображение графиков
plt.show()
```

# Краевая задача
$
\begin{align*}
    u'' - u' &=\cosh(2x)\\
        u(a) &= 0 \\
        u(b) + u'(b) &= 0 \\
        x &\in [0,10]
\end{align*}
$

# Конечно-разностный метод

```python
n = [11, 101, 1001, 10001]
d = [[0] * len(n)]

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs = axs.ravel()

for i in range(len(n)):
    data2 = np.loadtxt("X4_out.csv")
    xt1 = data2[:n[i], 0]
    x1 = data2[:n[i], 1]
    
    axs[i].plot(xt1, x1, label=f'u(x), h={10/(n[i]-1)}')
    axs[i].set_title(f'h={10/(n[i]-1)}')
    axs[i].legend(loc='lower left')
    
    with open("X4_out.csv","r") as f:
        lines = f.readlines()
    with open("X4_out.csv","w") as f:
        f.writelines(lines[n[i]:])  

import numpy as np
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt

def ode(x, y):
    return np.vstack((y[1], np.cosh(2*x) + y[1]))

def bc(ya, yb):
    return np.array([ya[0], yb[0]+yb[1]])

a = 0
b = 1
x = np.linspace(a, b, 10001)
y = np.zeros((2, x.size))

sol = solve_bvp(ode, bc, x, y)

x_plot = np.linspace(a, b, 10001)
y_plot = sol.sol(x_plot)[0]

plt.plot(x_plot, y_plot, label='u(x)')
plt.xlabel('x')
plt.ylabel('u(x)')
plt.legend()
plt.show()
# Отображение графиков
plt.show()
```
